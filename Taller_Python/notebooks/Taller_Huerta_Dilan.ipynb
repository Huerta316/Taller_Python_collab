Listas

```python
calif = [4, 6, 9, 7, 5]

suma = 0
for i in range(len(calif)):
  #print(f'Posicion {i} Calificacion: {calif[i]}')
  suma = suma + calif[i]

prom = suma / len(calif)
print(prom)

op = 0
for e in calif:
  op += e

print(f'Promedio: {op/len(calif)}')
```

Actividad 2. Dada una secuencia de 15 numeros, imprimir cantidad de numeros pares, convertir los numeros impares en pares, e imprimir la lista completa.

```python
numeros = []

for i in range(15):
    n = int(input(f"Ingresa el número {i+1}: "))
    numeros.append(n)

cantidad_pares = sum(1 for n in numeros if n % 2 == 0)

numeros_convertidos = [n if n % 2 == 0 else n + 1 for n in numeros]

print("\nCantidad de números pares:", cantidad_pares)
print("Lista completa con impares convertidos a pares:")
print(numeros_convertidos)
```

```python
import pandas as pd
import numpy as np

df = pd.read_excel('datos.xlsx')
df.head()

df['prom'] = df[['Parcial 1', 'Parcial 2', 'Parcial 3']].mean(axis=1)
df['estatus'] = np.where(df['prom'] >= 7, 'Aprobado', 'Reprobado')

df.to_excel('datos_mod.xlsx', index=False)
```
